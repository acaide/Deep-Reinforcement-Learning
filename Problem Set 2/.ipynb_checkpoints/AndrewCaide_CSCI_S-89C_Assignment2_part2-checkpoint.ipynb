### Part 1: Using classes Environment and Agent, implement the Gradient Bandit Algorithm with  ð›¼  = 0.4. Initialize the numerical preference at 0 for all actions.

In this section I develop the code. It isn't run until part 2.

```python
import random
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 

# Recycling from Problem Set 1

class Environment:
    def __init__(self):
        self.time = 1
        self.valid_actions = {1:4, 2:6, 3:8, 4:10} 
        
    def admissible_actions(self):
        return list(range(1,5))
    
    def get_reward(self, action):
        self.time += 1
        return random.gauss(self.valid_actions[action], 1)
```

```python
# Partially recycled from PS1
# Part 1:
class Agent:
    def __init__(self):
        self.current_reward = 0.0
        self.current_action = 0
        self.mean_reward = 0.0
        self.t = 0
        self.alpha = 0 
        self.history = pd.DataFrame(columns = ("Action", "Hits", "Preference", "Probability"), dtype = float)
        self.history = self.history.set_index("Action")

    def set_alpha(self, alpha):
        self.alpha = alpha
        
    def set_baseline(self, baseline):
        self.mean_reward = baseline
    
    def update_mean_reward(self):
        self.t += 1
        self.mean_reward = ((self.mean_reward * (self.t-1)) + self.current_reward)/self.t
        
        
    def update_probability(self):
        # Update all policies/probabilities
        # p_a = exp(h_a)/sum(exp(h_b))
        self.history["Probability"] = np.exp(self.history['Preference']) / \
            (np.exp(self.history['Preference']).sum())
        
    def update_action_preferences(self):
        
        # Update preference of current action:
        # H_(a) = h_a + alpha*(current_reward - mean_reward)(1-probability_a) 
        action_index = self.history.index.values == self.current_action
        action_preference = self.history.loc[action_index,"Preference"] +\
            self.alpha * (self.current_reward - self.mean_reward) * (1 - self.history.loc[action_index,"Probability"])
        self.history.loc[action_index,"Preference"] = action_preference
        
        # Update preferences of other actions:
        # H_a = h_a - alpha*(current_reward - mean_reward)*probability_a
        inactive_index = self.history.index.values != self.current_action
        inactive_preferences = self.history.loc[inactive_index,"Preference"] -\
            self.alpha * (self.current_reward - self.mean_reward) * self.history.loc[inactive_index,"Probability"]
        self.history.loc[inactive_index,"Preference"] = inactive_preferences 
        
    def take_action(self):
        self.current_action =  np.random.choice(self.history.index.values, 
                                                1, 
                                                p=self.history.Probability.values)[0]

        
    def step(self, env):
        actions = env.admissible_actions()
        
        # Initiate all possible actions to Ht = 0
        if(len(self.history.index) == 0): 
            for action in actions:
                self.history.loc[action,:] = [0,0,0]
            self.update_probability()
            self.take_action()
            self.current_reward = env.get_reward(self.current_action)
            self.update_mean_reward()
            self.update_action_preferences()
            self.update_probability()
            
        # Find max preference 
        else:
            
            # This automatically selects the action with the highest probability - bad implementation
            #max_h = self.history[self.history.Probability == self.history.Probability.max()]
            ## If there are ties in the, we pick a random one
            #if max_h.size > 1:
            #    i = random.choice(max_h.index.values)
            #else: 
            #    i = max_h.index.values   
            #self.current_action = i
            self.take_action()
            self.current_reward = env.get_reward(self.current_action)
            self.update_mean_reward()
            self.update_action_preferences()
            self.update_probability()
```

```python
# Run trials, keep track of actions, rewards
def rewards_hist(T = 100, baseline = 0, alpha = 0.4):
    env = Environment()
    agent = Agent()
    agent.set_alpha(alpha)
    agent.set_baseline(baseline)

    behavior = pd.DataFrame(columns = ("Action", "Reward"))
    while env.time <= T:
        agent.step(env)
        behavior = behavior.append({"Action":agent.current_action,
                           "Reward":agent.current_reward},
                           ignore_index = True)
    return behavior
```

### Part 2: Run the algorithm 500 times (i.e. for 500 episods) for T = 1000 and compute the observed average reward for each time  ð‘¡ .

```python
# Part 2
# Run epochs, keep track of actions, rewards, epochs, and time

def run_e(alpha = 0.4, epochs = 500, T = 100, baseline = 0):
    
    epoch_results = pd.DataFrame()
    for epoch in list(range(epochs)):
        
        run = rewards_hist(T, baseline, alpha)
        run['Epoch'] = epoch
        run['T'] = run.index
        epoch_results = epoch_results.append(run, ignore_index = True)
        if epoch % 100 == 0:
            print("Finished epoch {}".format(epoch))
    return epoch_results
```

```python
def test_multiple_alphas(alphas = [0,.1,.4,.8,1], epochs = 101, T = 500, baseline = 0):
    alpha_result = pd.DataFrame()
    for alpha in alphas:
        print("---\nAlpha: {}".format(alpha))
        alpha_run = run_e(alpha, epochs, T, baseline)
        aggregated = pd.DataFrame(alpha_run.groupby('T')['Reward'].mean())
        aggregated['Alpha'] = str(alpha)
        aggregated['T'] = aggregated.index
        alpha_result = alpha_result.append(aggregated, ignore_index = True)
    return(alpha_result)
```

### Part 1: Adding "state" to the Environment

```python
class Environment:
    def __init__(self):
        self.time = 1
        self.valid_actions = {1:4, 2:6, 3:8, 4:10} 
        self.current_state = 1
        
    def admissible_actions(self):
        return list(range(1,5))
    
    def state_change(self):
        return random.uniform(0, 1)>=0.95
    
    def state(self):
        if self.current_state == 1:
            self.valid_actions = {1:4, 2:6, 3:8, 4:10} 
        else:
            self.valid_actions = {1:-10, 2:-8, 3:-6, 4:-4} 

    def get_state(self):
        return(self.current_state)
    
    def get_reward(self, action):
        self.time += 1
        if action == 1 & self.state_change():
            print("State change detected!")
            self.current_state = 2
            self.state()
        return random.gauss(self.valid_actions[action], 1)
```

### Part 2: Modify the Gradient Bandit Algorithm - Track State 

Note: $\mathscr{S}$ X $\mathscr{A}$ $\rightarrow$ $\mathbb{R}$

This means rewards should be dependent on states, as should the probabilities

```python
# Part 2:
class Agent:
    def __init__(self):
        self.current_reward = 0.0
        self.current_action = 0
        self.mean_reward = 0.0
        self.t = 0
        self.alpha = 0 
        self.history = pd.DataFrame(columns = ("State","Action", "Hits", "Preference", "Probability"), dtype = float)
        self.history = self.history.set_index(["State","Action"])
        self.state = 1

    def set_alpha(self, alpha):
        self.alpha = alpha
        
    def set_baseline(self, baseline):
        self.mean_reward = baseline
    
    def update_mean_reward(self):
        self.t += 1
        self.mean_reward = ((self.mean_reward * (self.t-1)) + self.current_reward)/self.t
        
        
    def update_probability(self):
        # Keep track of state
        stateloc = self.history.loc[(self.state)] 
        
        # Update all policies/probabilities
        # p_a = exp(h_a)/sum(exp(h_b))
        stateloc["Probability"] = np.exp(stateloc['Preference']) / \
            (np.exp(stateloc['Preference']).sum())
        
    def update_action_preferences(self):

        # Keep track of state
        stateloc = self.history.loc[(self.state)] 
        
        # Update preference of current action:
        # H_(a) = h_a + alpha*(current_reward - mean_reward)(1-probability_a)
        action_index = stateloc.index.values == self.current_action
        action_preference = stateloc.loc[action_index,"Preference"] +\
            self.alpha * (self.current_reward - self.mean_reward) * (1 - stateloc.loc[action_index,"Probability"])
        stateloc.loc[action_index,"Preference"] = action_preference
        
        # Update preferences of other actions:
        # H_a = h_a - alpha*(current_reward - mean_reward)*probability_a
        inactive_index = stateloc.index.values != self.current_action
        inactive_preferences = stateloc.loc[inactive_index,"Preference"] -\
            self.alpha * (self.current_reward - self.mean_reward) * stateloc.loc[inactive_index,"Probability"]
        stateloc.loc[inactive_index,"Preference"] = inactive_preferences 
        
    def take_action(self):
        stateloc = self.history.loc[(self.state)] 
        self.current_action =  np.random.choice(stateloc.index.values, 1, 
                                                p=stateloc.Probability.values)[0]
     
    # --------------------------------
    # This is how my algorithm works:    
    def sequence(self, env):
        self.take_action() # Make a move
        self.current_reward = env.get_reward(self.current_action) # Get the reward
        self.state = env.get_state() # Get the state
        self.update_mean_reward() # Update the reward of the action just taken
        self.update_action_preferences() # Update action preferences based on the actions just taken
        self.update_probability() # Update probabilities
    # --------------------------------

    def step(self, env):
        actions = env.admissible_actions()

        if(len(self.history.index) == 0): 
            for state in [1,2]:
                for action in actions:
                    self.history.loc[(state,action),:] = [0,0,0]
            self.update_probability()
            self.sequence(env)
        else:
            self.sequence(env)
```

```python
# Run trials, keep track of actions, rewards
def rewards_hist(T = 100, baseline = 0, alpha = 0.4):
    env = Environment()
    agent = Agent()
    agent.set_alpha(alpha)
    agent.set_baseline(baseline)

    behavior = pd.DataFrame(columns = ("Action", "Reward"))
    while env.time <= T:
        agent.step(env)
        behavior = behavior.append({"Action":agent.current_action,
                                    "State":agent.state,
                                    "Reward":agent.current_reward},
                                    ignore_index = True)
    return behavior, agent.history

def run_e(alpha = 0.4, epochs = 500, T = 100, baseline = 0):
    
    epoch_results = pd.DataFrame()
    for epoch in list(range(epochs)):
        epoch_run = rewards_hist(T, baseline, alpha)
        run = epoch_run[0]
        run['Epoch'] = epoch
        run['T'] = run.index
        epoch_results = epoch_results.append(run, ignore_index = True)
        if epoch % 100 == 0:
            print("Finished epoch {}".format(epoch))
    return epoch_results, epoch_run[1]
```

### Does the Gradient Bandit Algorithm appear to be appropriate in the case of Markov Decision Process (MDP)? Explain

In this specific problem, no. This is because of how the problem is set up: If the GBA mistakenly selects A1 in S1 (mistakenly becasue it's the lowest scoring action) there's a 5% chance it will be thrown into the penalty state, S2. 

Once in S2, the only way to escape back to S1 is to select the most damaging action, A1 (-10 points) with a 5% chance of escaping back up to S1. A1 (-10 points) is 2.5x worse than selecting A4 (-4), therefore the GBA is disincentivized from experimenting too much, especially at a state-switching rate of 5%.
